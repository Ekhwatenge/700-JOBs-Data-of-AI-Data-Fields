```python
import pandas as pd
import matplotlib.pyplot as plt

def quick_visualize_all(df):
    """Create all visualizations at once"""
    
    # Job types
    job_type_cols = ['jobType', 'job_type', 'type', 'category']
    for col in job_type_cols:
        if col in df.columns:
            plt.figure(figsize=(10, 6))
            df[col].value_counts().head(8).plot(kind='pie', autopct='%1.1f%%')
            plt.title(f'Job Type Distribution ({col})')
            plt.show()
            break
    
    # Top companies
    company_cols = ['company', 'companyName', 'company_name']
    for col in company_cols:
        if col in df.columns:
            plt.figure(figsize=(12, 8))
            df[col].value_counts().head(15).plot(kind='barh')
            plt.title('Top Companies')
            plt.show()
            break
    
    # Locations
    location_cols = ['location', 'city', 'state']
    for col in location_cols:
        if col in df.columns:
            plt.figure(figsize=(12, 6))
            df[col].value_counts().head(10).plot(kind='bar')
            plt.title('Top Locations')
            plt.xticks(rotation=45)
            plt.show()
            break

# Load and visualize
df = pd.read_csv('jobs_dataset.csv')
quick_visualize_all(df)
```

```python
#create one chart at a time
df = pd.read_csv('jobs_dataset.csv')

def plot_job_type_distribution(df):
    """Create pie chart for job types"""
    job_type_cols = ['jobType', 'job_type', 'type', 'category', 'employment_type']
    
    for col in job_type_cols:
        if col in df.columns:
            plt.figure(figsize=(10, 8))
            job_counts = df[col].value_counts().head(8)
            plt.pie(job_counts.values, labels=job_counts.index, autopct='%1.1f%%', startangle=90)
            plt.title(f'Job Type Distribution ({col})', fontsize=16, fontweight='bold')
            plt.axis('equal')
            plt.show()
            return

def plot_top_companies(df):
    """Create horizontal bar chart for top companies"""
    company_cols = ['company', 'companyName', 'company_name', 'employer']
    
    for col in company_cols:
        if col in df.columns:
            plt.figure(figsize=(12, 8))
            top_companies = df[col].value_counts().head(15)
            plt.barh(range(len(top_companies)), top_companies.values, color='skyblue')
            plt.yticks(range(len(top_companies)), top_companies.index)
            plt.xlabel('Number of Jobs')
            plt.title('Top 15 Companies by Job Postings', fontsize=16, fontweight='bold')
            plt.gca().invert_yaxis()
            plt.tight_layout()
            plt.show()
            return

# Usage:
plot_job_type_distribution(df)
plot_top_companies(df)
```
